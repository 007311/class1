```julia
using Logging

logger = ConsoleLogger()
function phi1(n)::BigFloat
    n >= 1 || return 0
    n == 1 && return 1
    1 + 1/phi1(n-1)
end

function phi2(n)::BigFloat
    n >= 1 || return 0
    n == 1 && return 1
    sqrt(1 + phi2(n-1))
end

function factorial(n)::BigInt
    n >= 0 || error("n must be non-negative")
    n == 0 && return 1
    n * factorial(n-1)
end

function fib1(n)::Pair{BigInt, BigInt}
    n >= 1 || return 0
    n == 1 && return Pair(0,1)
    fm = fib1(n-1)
    #@info "fm is: " fm
    return Pair(fm.second, fm.first + fm.second)
end

function fib(n)::Tuple{BigInt, BigInt}
    n >= 0 || error("n must be non-negative")
    n >= 1 || return 0,0
    n == 1 && return 0,1
    fm = fib(n-1)
    @info "fm is: " fm
    return fm[2], fm[1] + fm[2]
end

function e()::BigFloat
    e = 0
    for x = 0:62
        e = e + 1/factorial(x)
    end
    return e
end

with_logger(logger) do
    fibx = fib(BigInt(200))
    phix = BigFloat(fibx[1] / fibx[2])
    println(phix)
end

#println(e())
#println(BigFloat(ℯ))
y = BigFloat(185)
#z = BigFloat(154)
println(phi1(y))
#x = BigFloat(5)
#println((sqrt(x)+1)/2)
#println(phi2(z))       
```
